```python
# the project is to analyse data on the companies that are on Dividend aristocrat list by The Motley Fool
import requests
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://www.fool.com/investing/what-is-a-dividend-aristocrat.aspx'
html = requests.get(url).content #Import data from the Url
df_list = pd.read_html(html)  #Conversion to the list
df = df_list[0]
df.to_csv('my data.csv')  #Conversion of the list to csv files
df
```

```python
df1=df[df['Consecutive Years of Dividend Growth']>40] #companies with dividend growth for more than 40years 
df1
```

```python
df.plot(x ='Company', y='Consecutive Years of Dividend Growth', kind = 'scatter')
plt.show()
```

```python
df1.plot(x ='Company', y='Consecutive Years of Dividend Growth', kind = 'bar')
plt.show()
```

```python
df2=df[df['Sector']=='Healthcare']
dff=df2.sort_values('Consecutive Years of Dividend Growth')
dff.mean()
```

```python
#showing only the healthcare stock
dff.plot(x ='Company', y='Consecutive Years of Dividend Growth', kind = 'bar')
plt.show()
```

```python
print(df[df['Sector']=='Healthcare'].count()) #counting number of companies in each sector
print(df[df['Sector']=='Consumer staples'].count()
print(df[df['Sector']=='Consumer discretionary'].count())
print(df[df['Sector']=='Industrials'].count())

```

```python
sectors=df.groupby('Sector').count()  # grouping each sectors, easier and efficient code than line 15
sectors 
```

```python
sectors=sectors.reset_index(0) #applyig index to sectors
sectors
```

```python
#Showing the number of top dividend paying stocks in each sectors
sectors.plot(x ='Sector', y='Company', kind = 'bar')
plt.show()
```
